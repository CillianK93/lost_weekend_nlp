```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the data from the excel files.
lost_reviews = pd.read_excel("review_data/lost_reviews.xlsx")
reviews_df = pd.read_excel("review_data/reviews_df.xlsx")
reviews_df_2 = pd.read_excel("review_data/reviews_df_2.xlsx")
reviews_df_3 = pd.read_excel("review_data/reviews_df_3.xlsx")
```

```python
# Concat the dfs together
reviews_df = pd.concat([lost_reviews, reviews_df, reviews_df_2, reviews_df_3], ignore_index = True)
pd.set_option('display.max_columns', None)
reviews_df
```

# Data Cleaning
- There is a lot of things that can be optimised from this raw data such as unneeded columns, NaN values and formatting

```python
reviews_df.columns
```

```python
# removing columns not needed
# the last 26 columns were not needed
reviews_df = reviews_df.iloc[:, :-26]
```

```python
# These columns will also not be needed for the project
other_columns = ['query', 'google_id', 'place_id', 'location_link', 'reviews_link', 
              'reviews', 'review_id', 'author_link', 'author_id', 'author_image', 'review_img_url',
              'review_img_urls', 'review_photo_ids', 'owner_answer', 'owner_answer_timestamp', 
              'owner_answer_timestamp_datetime_utc','review_link','review_timestamp']

reviews_df = reviews_df.drop(columns = other_columns)

reviews_df
```

```python
# unfortunately due to the nature of leaving reviews not all people write comments and just leave a rating
# So I will need to remove these reviews :(
reviews_df.isna().sum()
```

```python
# upon inspection we have a row of all NaNs i will drop this
reviews_df = reviews_df.dropna(how = 'all')
```

### Seeing as we scraped 350 reviews from all business, dropping NaNs might affect one business more than the other so i will visualize this to see where most reviewers did not leave a comment.

```python
# creating a df to count the nans for each venue.
nan_counts_per_venue = reviews_df.groupby('name')['review_text'].apply(lambda x: x.isna().sum())
```

```python
reviews_df['name'].value_counts()
```

```python
# Visualising using a barplot


plt.figure(figsize=(12, 8))

nan_counts_per_venue.sort_values().plot(kind='barh', color='skyblue')

plt.title('Number of NaNs per Venue')
plt.xlabel('Number of NaNs')
plt.ylabel('Venue')

plt.tight_layout()
plt.show()
```

- I also wonder if the NaN values of review_text will differentiate from the rating of the review that was left.

```python
nans_by_rating = reviews_df.groupby('review_rating')['review_text'].apply(lambda x: x.isna().sum())
```

```python
plt.figure(figsize=(12,8))

nans_by_rating.plot(kind='bar', color='skyblue')

plt.title('Number of NaNs in Reviews Text per Rating')
plt.xlabel('Rating')
plt.ylabel('Number of NaNs')

plt.tight_layout()
plt.show()

```

#### Interesting to see a high amount of NaNs in the 5 star reviews.
- People arent as vocal if they really like the venue and but will still leave a review
- If people arent happy they are more likely to leave a comment on why.

```python
# dropping NaNs in review_text and reseting the index

reviews_df = reviews_df.dropna().reset_index(drop = True)
```

```python
reviews_df
```

```python
# I've noticed the dates are in the month/day format and would like to change to day/month
reviews_df.info()
```

```python
reviews_df['extracted_date'] = reviews_df['review_datetime_utc'].str.extract(r'(\d{2}/\d{2}/\d{4})')
reviews_df
```

```python
# Example: Convert 'MM/DD/YYYY' to 'DD/MM/YYYY' format

# Using string slicing to break the date into components
month = reviews_df['extracted_date'].str.slice(0, 2)
day = reviews_df['extracted_date'].str.slice(3, 5)
year = reviews_df['extracted_date'].str.slice(6, 10)

# Reassembling the date.
corrected_date = day + '/' + month + '/' + year

# Assign corrected dates back to the DataFrame
reviews_df['date'] = corrected_date
```

```python
reviews_df = reviews_df.drop(columns = ['review_datetime_utc', 'extracted_date'])
```
